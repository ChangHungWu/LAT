```python
import tkinter as tk
import requests

keyfile = open("key.txt", "r")
key = keyfile.readline()
key = keyfile.readline()

api_key = key
location = "25.013413,121.540475"

def calculate_distance(lat1, lon1, lat2, lon2):
    """
    使用 Haversine 公式計算兩組座標之間的距離。
    """
    from math import radians, sin, cos, sqrt, atan2

    # 地球半徑，單位為公里
    R = 6371.0

    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)

    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad

    a = sin(dlat / 2) ** 2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

def search_nearby_places():
    keyword = keyword_entry.get()  # 從使用者輸入中獲取關鍵字

    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    params = {
        "key": api_key,
        "location": location,
        "radius": "2000",
        "keyword": keyword,
        "language": "zh-TW"
    }

    response = requests.get(url, params=params)
    data = response.json()

    results_listbox.delete(0, tk.END)  # 清除先前的搜尋結果

    if data["status"] == "OK":
        results = data["results"]
        landmark_lat, landmark_lon = map(float, location.split(","))
        
        # 建立一個列表來存儲帶有距離的搜尋結果
        results_with_distance = []
        
        for place in results:
            name = place["name"]
            address_components = place["plus_code"]["compound_code"].split()
            county = [c for c in address_components if "縣" in c or "市" in c][0]
            address = place["vicinity"].replace(county, "")
            place_lat = place["geometry"]["location"]["lat"]
            place_lon = place["geometry"]["location"]["lng"]
            distance = calculate_distance(landmark_lat, landmark_lon, place_lat, place_lon)
            rating = place.get("rating", "N/A")  # 檢查是否有評分，若無則設為 "N/A"
            
            # 將帶有距離的搜尋結果加入列表
            results_with_distance.append((name, address, distance, rating))
        
        # 根據距離對結果進行排序
        sorted_results = sorted(results_with_distance, key=lambda x: x[2])
        
        for result in sorted_results:
            name = result[0]
            address = result[1]
            distance = result[2]
            rating = result[3]
            
            # 在結果列表中顯示店名、地址、距離和評分
            results_listbox.insert(tk.END, f"店名: {name}\n")
            results_listbox.insert(tk.END, f"地址: {address}\n")
            results_listbox.insert(tk.END, f"距離: {distance:.2f} 公里\n")
            results_listbox.insert(tk.END, f"評分: {rating}\n")
            results_listbox.insert(tk.END, "\n")
    else:
        results_listbox.insert(tk.END, "沒有找到結果。")

# 建立使用者介面
root = tk.Tk()
root.title("附近地點搜尋")

# 建立位置和關鍵字輸入欄位
location_label = tk.Label(root, text="位置:")
location_label.grid(row=0, column=0, padx=5, pady=5)
location_entry = tk.Entry(root, width=40)
location_entry.grid(row=0, column=1, padx=5, pady=5)
location_entry.insert(0, "國立臺灣師範大學")

keyword_label = tk.Label(root, text="關鍵字:")
keyword_label.grid(row=1, column=0, padx=5, pady=5)
keyword_entry = tk.Entry(root, width=40)
keyword_entry.grid(row=1, column=1, padx=5, pady=5)
keyword_entry.insert(0, "想找什麼呢")

# 建立搜尋按鈕
search_button = tk.Button(root, text="搜尋", command=search_nearby_places)
search_button.grid(row=2, column=0, columnspan=2, padx=5, pady=5)

# 建立結果列表框架
results_frame = tk.Frame(root)
results_frame.grid(row=3, column=0, columnspan=2, padx=5, pady=5)

# 建立結果列表
results_listbox = tk.Listbox(results_frame, width=60, height=15)
results_listbox.pack(side=tk.LEFT, fill=tk.BOTH)
results_scrollbar = tk.Scrollbar(results_frame, orient=tk.VERTICAL, command=results_listbox.yview)
results_scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
results_listbox.config(yscrollcommand=results_scrollbar.set)

# 啟動使用者介面
root.mainloop()
```
