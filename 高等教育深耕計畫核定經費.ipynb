```python
import pandas as pd

df110 = pd.read_csv('高等教育深耕計畫核定經費110年 .csv')

df110
```

```python
temp = df110[['類別','學校名稱', '核定經費(單位：元)']]

temp
```

# 區分110年一般大學和技術學院的總經費

```python
new['year'] = 110
new
```
