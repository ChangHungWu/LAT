```python
#把第一個CSV讀入
import pandas as pd

df = pd.read_csv('高等教育深耕計畫核定經費110年.csv',encoding = 'big5')

df
```

```python
temp = df[['學校名稱', '核定經費(單位：元)']]
temp
```

```python
Type = temp.groupby(['學校名稱'])['核定經費(單位：元)'].sum().reset_index()
```

```python
Type['year'] = 110
Type
```

```python
df107 = pd.read_csv('高等教育深耕計畫核定經費107年.csv',encoding = 'big5')
df108 = pd.read_csv('高等教育深耕計畫核定經費108年.csv',encoding = 'big5')
df109 = pd.read_csv('高等教育深耕計畫核定經費109年.csv',encoding = 'big5')
```

```python
new2 = df108[['學校名稱', '核定經費(單位：元)']].groupby(['學校名稱'])['核定經費(單位：元)'].sum().reset_index()
new2['year'] = 108
new2
```

```python
new3 = df109[['學校名稱', '核定經費(單位：元)']].groupby(['學校名稱'])['核定經費(單位：元)'].sum().reset_index()
new3['year'] = 109
new3
```
